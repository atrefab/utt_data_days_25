# Installation

1. Cloner le dépôt: git clone https://github.com/atrefab/utt_data_days_25.git
2. Exécuter setup.bat pour installer l'environnement Python

# Démarrage des services

1. Exécuter activate.bat pour activer l'environnement dans un nouveau terminal
2. Exécuter start_liquid.bat pour lancer Liquid
3. Exécuter start_mistral.bat pour lancer Mistral
4. Exécuter start_jupyter pour lancer jupyterlab
